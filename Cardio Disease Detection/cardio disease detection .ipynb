```python
#outlier treatment
```

```python
q3 = data['height'].quantile(0.75)
q1 = data['height'].quantile(0.25)
iqr = q3 - q1
```

```python
ulo = q3 + 1.5*(iqr)
ulo
```

```python
llo = q1 - 1.5*(iqr)
llo
```

```python
outlier_data = data[(data['height'] >= ulo) | (data['height'] <= llo)]
outlier_data
```

```python
data.shape
```

```python
outlier_data['height'].value_counts().index.shape
```

```python
x = data.iloc[:,:-1]
x.shape
```

```python
y = data.iloc[:, -1]
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = True)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression, LinearRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
pred_lr = lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
accuracy_score(pred_lr, y_test)
```

```python
print(classification_report(pred_lr, y_test))
```

```python
confusion_matrix(pred_lr, y_test)
```

```python
err = []
for a in range(1, 40, 2):
    model = KNeighborsClassifier(n_neighbors = a)
    model.fit(x_train, y_train)
    y_pred_a = model.predict(x_test)
    
    err.append(np.mean(y_pred_a != y_test))
```

```python
plt.figure(figsize = (19, 8))
plt.plot(range(1, 40, 2), err)
plt.xlabel('Kvalues')
plt.ylabel('Error')
plt.title('Optimum K Value')
```
